```python
import re

logs = full_data.strip().split('\n')

depth = 0
horizontal = 0

for l in logs:
    g = re.match(r"(\w+) (\w+)", l)
    if g.group(1) == "up":
        depth -= int(g.group(2))
    elif g.group(1) == "down":
        depth += int(g.group(2))
    else:
        horizontal += int(g.group(2))


ans = depth * horizontal
print(f"ans is {ans}")
```
